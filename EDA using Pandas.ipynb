```python
# read data
import torch
import pandas as pd
import numpy as np
import os
import math
from tqdm import tqdm
```

```python
# use regex to remove links from tweets
import re

extractedTweets_df['Tweet'] = extractedTweets_df['Tweet'].apply(lambda Tweet: re.sub(r"(?:http|pic.twitter)\S+", "", Tweet))
```

```python
# remove new line, @, #, and RT
extractedTweets_df['Tweet'] = extractedTweets_df['Tweet'].apply(lambda Tweet: re.sub("\n"," ",Tweet))
extractedTweets_df['Tweet'] = extractedTweets_df['Tweet'].apply(lambda Tweet: re.sub("@","",Tweet))
extractedTweets_df['Tweet'] = extractedTweets_df['Tweet'].apply(lambda Tweet: re.sub("RT","",Tweet))
extractedTweets_df['Tweet'] = extractedTweets_df['Tweet'].apply(lambda Tweet: re.sub("#","",Tweet))
```

```python
import string

# remove punctuations and non-ascii characters
printable = set(string.printable)

def remove_non_ascii(s):
  filtered = list(filter(lambda x: x in printable, s))
  filtered = ''.join(filtered)
  filtered = re.sub("[^\w\d'\s]+",'',filtered)
  return filtered

extractedTweets_df['Tweet'] = extractedTweets_df['Tweet'].apply(lambda Tweet: remove_non_ascii(Tweet))
```
